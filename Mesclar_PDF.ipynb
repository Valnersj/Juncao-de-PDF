```python
from PyPDF2 import PdfMerger
from google.colab import files
```

```python

# 1. Selecionar arquivos para realizar a junção
files.upload()

# 2. Mesclando os PDFs selecionados
merger = PdfMerger()
merger.append("Boleto_MBA_11_2025.pdf")
merger.append("ComprovanteSantander - Novembro_2025.pdf")

merger.write("pdf_boleto_comprovante_Novembro_2025.pdf")
merger.close()

# 3. Download do resultado final
files.download("pdf_boleto_comprovante_Novembro_2025.pdf")
```
